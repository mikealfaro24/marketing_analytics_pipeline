# Silver Layer — Customer Demographics
Clean customer demographics data from Bronze and write standardized Silver table.

## 1. Load bronze table

```python
# Load Bronze table
df = spark.table("shoply_analytics.bronze.customer_demographics")
```

## 2. Deduplicate user_id

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number, col

# Partition by user_id and keep first row
window = Window.partitionBy("user_id").orderBy(col("user_id"))

df_clean = (
    df.withColumn("rn", row_number().over(window))
      .filter("rn = 1")
      .drop("rn")
)
```

## 3. Write to Silver

```python
df_clean.write.format("delta").mode("overwrite").saveAsTable( "shoply_analytics.silver.customer_demographics" )
```
