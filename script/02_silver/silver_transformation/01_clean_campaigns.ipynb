# Silver Layer — Campaigns
Clean campaign data from Bronze and write standardized Silver table.


```python
# read data from bronze table
df = spark.table("shoply_analytics.bronze.campaigns")
```

```python
from pyspark.sql.functions import col, trim, when

df_clean = (
    df.withColumn(
        "budget",
        when(trim(col("budget")) == "", None)  # convert empty/space-only strings to NULL
        .otherwise(col("budget").cast("double"))
    )
)

```

```python
df_clean.write.format("delta").mode("overwrite").saveAsTable( "shoply_analytics.silver.campaigns" )
```
