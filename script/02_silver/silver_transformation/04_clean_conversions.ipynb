# Silver Layer — Conversions
Clean conversion data from Bronze and write standardized Silver table.

## 1. Load bronze table

```python
# Load Bronze table
df = spark.table("shoply_analytics.bronze.conversions")
```

## 2. Deduplicate conversion_id

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number, col

window = Window.partitionBy("conversion_id").orderBy(col("timestamp").desc())

df_clean = (
    df.withColumn("rn", row_number().over(window))
      .filter("rn = 1")
      .drop("rn")
)
```

##  3. Load to silver

```python
df_clean.write.mode("overwrite").saveAsTable("shoply_analytics.silver.conversion")
```
