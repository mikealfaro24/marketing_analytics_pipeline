```sql
-- ============================================
-- QC — Customer Demographics (Bronze)
-- ============================================

USE CATALOG shoply_analytics;
USE SCHEMA bronze;

-- 1. Preview
SELECT *
FROM customer_demographics
LIMIT 20;

-- 2. Row count
-- 9866 rows
SELECT COUNT(*) AS row_count
FROM customer_demographics;

-- 3. Duplicate user_id
-- 2 duplicates 
SELECT user_id, COUNT(*) AS cnt
FROM customer_demographics
GROUP BY user_id
HAVING COUNT(*) > 1;

-- 4. Null checks
-- 610 null gender
SELECT
  SUM(CASE WHEN user_id IS NULL THEN 1 END) AS null_user_id,
  SUM(CASE WHEN age IS NULL THEN 1 END) AS null_age,
  SUM(CASE WHEN gender IS NULL THEN 1 END) AS null_gender,
  SUM(CASE WHEN country IS NULL THEN 1 END) AS null_country,
  SUM(CASE WHEN email IS NULL THEN 1 END) AS null_email
FROM customer_demographics;

-- 5. Age sanity checks
SELECT *
FROM customer_demographics
WHERE age < 0 OR age > 120;

-- 6. Unexpected gender values
SELECT DISTINCT gender
FROM customer_demographics;

-- 7. Unwanted spaces in string columns
SELECT *
FROM customer_demographics
WHERE user_id != TRIM(user_id)
   OR gender != TRIM(gender)
   OR country != TRIM(country)
   OR email != TRIM(email);

-- 8. Email format check
SELECT *
FROM customer_demographics
WHERE email NOT LIKE '%@%' AND email IS NOT NULL;

-- 9. Column structure
DESCRIBE TABLE customer_demographics;
```
