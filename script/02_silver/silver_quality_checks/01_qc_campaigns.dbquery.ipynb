```sql
-- ============================================
-- QC — Campaigns (Bronze)
-- ============================================

USE CATALOG shoply_analytics;
USE SCHEMA bronze;

-- issue log:
-- budget is a string 
-- 9 start date nulls
-- 223 end date nulls
-- 11  budget null


-- 1. Preview

SELECT *
FROM campaigns
LIMIT 20;

-- 2. Row count 
-- 820 
SELECT COUNT(*) AS row_count FROM campaigns;

-- 3. Duplicate campaign_id 
-- no results returned 
SELECT campaign_id, 
  COUNT(*) AS cnt
FROM campaigns 
GROUP BY campaign_id 
HAVING COUNT(*) > 1;

-- 4. Null checks 
-- 9 start date nulls
-- 223 end date nulls
-- 11  budget null
SELECT 
SUM(CASE WHEN campaign_id IS NULL THEN 1 END) AS null_campaign_id, 
SUM(CASE WHEN campaign_name IS NULL THEN 1 END) AS null_campaign_name, 
SUM(CASE WHEN source IS NULL THEN 1 END) AS null_source, 
SUM(CASE WHEN medium IS NULL THEN 1 END) AS null_medium, 
SUM(CASE WHEN start_date IS NULL THEN 1 END) AS null_start_date, SUM(CASE WHEN end_date IS NULL THEN 1 END) AS null_end_date, 
SUM(CASE WHEN budget IS NULL THEN 1 END) AS null_budget
FROM campaigns;

-- 5. Date validity (end_date < start_date) 
-- looks clean (handled in excel)
SELECT * 
FROM campaigns
WHERE end_date < start_date;

-- 6. Budget sanity check (negative or zero) 
-- budget is string 
-- 4 zero
SELECT * 
FROM campaigns 
WHERE try_cast(budget as float) <= 0;

-- 7. Unwanted spaces check (should be clean from Excel) 
SELECT *
 FROM campaigns 
 WHERE campaign_name != TRIM(campaign_name) 
        OR source != TRIM(source) OR medium != TRIM(medium);


-- 8. Column structure 
DESCRIBE TABLE campaigns;
```
