```sql
-- ============================================
-- QC — Conversions (Bronze)
-- ============================================

USE CATALOG shoply_analytics;
USE SCHEMA bronze;


-- 1. Preview
-- time contains in different format
SELECT *
FROM conversions
LIMIT 20;

-- 2. Row count
-- count: 9804
SELECT COUNT(*) AS row_count
FROM conversions;

-- 3. Duplicate conversion_id
-- 4 duplicates 

SELECT conversion_id, COUNT(*) AS cnt
FROM conversions
GROUP BY conversion_id
HAVING COUNT(*) > 1;

-- 4. Null checks
-- user_id 2733
-- 54 null amount 
-- 96 null timestamp
-- 4887 null campaign_id

SELECT
  SUM(CASE WHEN conversion_id IS NULL THEN 1 END) AS null_conversion_id,
  SUM(CASE WHEN session_id IS NULL THEN 1 END) AS null_session_id,
  SUM(CASE WHEN user_id IS NULL THEN 1 END) AS null_user_id,
  SUM(CASE WHEN conversion_type IS NULL THEN 1 END) AS null_conversion_type,
  SUM(CASE WHEN amount IS NULL THEN 1 END) AS null_amount,
  SUM(CASE WHEN timestamp IS NULL THEN 1 END) AS null_timestamp,
  SUM(CASE WHEN campaign_id IS NULL THEN 1 END) AS null_campaign_id
FROM conversions;

-- 5. Check for malformed timestamps
-- we had some malformed timestamps 1000 rows maybe more
SELECT *
FROM conversions
WHERE TRY_TO_TIMESTAMP(timestamp) IS NULL
  AND timestamp IS NOT NULL
  AND TRIM(timestamp) != '';

-- 6. Future timestamps
-- no rows returned 
SELECT *
FROM conversions
WHERE timestamp > current_timestamp();

-- 7. Revenue sanity check (negative or extreme values)
-- no rows return
SELECT *
FROM conversions
WHERE amount < 0;

-- 8. Unwanted spaces in conversion_type
-- no rows returned
SELECT *
FROM conversions
WHERE conversion_type != TRIM(conversion_type);

-- 9. Unexpected conversion_type values
-- expected values
SELECT DISTINCT conversion_type
FROM conversions;


-- 10. Column structure
DESCRIBE TABLE conversions;


```
