```sql
-- ============================================================
-- GOLD DIMENSION: dim_device
-- PURPOSE:
--   Store the unique device types used by website visitors.
--
-- PRIMARY KEY:
--   device_id
--
-- RELATIONSHIP:
--   fact_marketing_performance.device_id
--     -> dim_device.device_id
--
-- SOURCE:
--   shoply_analytics.silver.sessions
--
-- NOTE:
--   Missing or blank device values are mapped to 'unknown'
--   so the fact table has a valid foreign key target.
-- ============================================================

USE CATALOG shoply_analytics;
USE SCHEMA gold;

CREATE OR REPLACE VIEW dim_device AS
WITH distinct_devices AS (
    SELECT DISTINCT
        COALESCE(NULLIF(device_type, ''), 'unknown') AS device_type
    FROM shoply_analytics.silver.sessions
)

SELECT
    -- Surrogate primary key
    ROW_NUMBER() OVER (
        ORDER BY device_type
    ) AS device_id,

    -- Device attribute
    device_type

FROM distinct_devices;
```
