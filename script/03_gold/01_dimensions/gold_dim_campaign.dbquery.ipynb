```sql
-- ============================================================
-- GOLD DIMENSION: dim_campaign
-- PURPOSE:
--   Store descriptive information about each marketing campaign.
--
-- PRIMARY KEY:
--   campaign_id
--
-- RELATIONSHIP:
--   fact_marketing_performance.campaign_id
--     -> dim_campaign.campaign_id
--
-- SOURCE:
--   shoply_analytics.silver.campaigns
-- ============================================================

USE CATALOG shoply_analytics;
USE SCHEMA gold;

CREATE OR REPLACE VIEW dim_campaign AS
SELECT
    -- Primary key
    campaign_id,

    -- Descriptive attributes
    campaign_name,
    start_date,
    end_date,
    budget AS campaign_budget

FROM shoply_analytics.silver.campaigns;
```
