```sql
-- ============================================================
-- GOLD DIMENSION: dim_customer
-- PURPOSE:
--   Store customer demographic information for segmentation and
--   descriptive analysis.
--
-- PRIMARY KEY:
--   customer_id
--
-- RELATIONSHIP:
--   fact_marketing_performance.customer_id
--     -> dim_customer.customer_id
--
-- SOURCE:
--   shoply_analytics.silver.customer_demographics
--
-- NOTE:
--   customer_id is aligned directly to user_id from the
--   customer demographics dataset.
-- ============================================================

USE CATALOG shoply_analytics;
USE SCHEMA gold;

CREATE OR REPLACE VIEW dim_customer AS
SELECT
    -- Primary key
    user_id AS customer_id,

    -- Descriptive attributes
    first_name,
    last_name,
    email,
    age,

    -- Derived age band for easier segmentation
    CASE
        WHEN age IS NULL THEN 'unknown'
        WHEN age < 18 THEN 'under_18'
        WHEN age BETWEEN 18 AND 24 THEN '18_24'
        WHEN age BETWEEN 25 AND 34 THEN '25_34'
        WHEN age BETWEEN 35 AND 44 THEN '35_44'
        WHEN age BETWEEN 45 AND 54 THEN '45_54'
        WHEN age BETWEEN 55 AND 64 THEN '55_64'
        ELSE '65_plus'
    END AS age_group,

    gender,
    country,
    signup_date,
    marketing_opt_in

FROM shoply_analytics.silver.customer_demographics;
```
